```python
df_hugging = pd.read_csv(
    "companyDataClean.csv"
)
df_hugging
```

```python
df_hugging["HSNR"] = df_hugging["STREET"].apply(extract_house_number)
df_hugging["STREET"] = df_hugging["STREET"].apply(extract_street_name)
df_hugging
```
